```python
# Step 3: For each start and end time, find closest matching timestamp in location_df
def get_closest_location(device_id, timestamp, _location_df ):
    subset = _location_df[_location_df['device_id'] == device_id]
    
    if subset.empty:
        print(f"⚠️ No location data found for device_id {device_id}")
        return pd.Series({'x': None, 'y': None, 'location_time': None})
    
    subset['time_diff'] = (subset['date_created'] - timestamp).abs()
    
    if subset['time_diff'].isna().all():
        print(f"⚠️ All time differences are NaT for device_id {device_id} and time {timestamp}")
        return pd.Series({'x': None, 'y': None, 'location_time': None})
    
    closest = subset.loc[subset['time_diff'].idxmin()]
    return pd.Series({'x': closest['x'], 'y': closest['y'], 'location_time': closest['date_created']})

paired_trips2[['start_x', 'start_y', 'start_loc_time']] = paired_trips2.apply(
    lambda row: get_closest_location(row['device_id'], row['start_time'], loc_df2), axis=1
)

paired_trips2[['end_x', 'end_y', 'end_loc_time']] = paired_trips2.apply(
    lambda row: get_closest_location(row['device_id'], row['end_time'], loc_df2), axis=1
)

# Result: DataFrame with trips and their start/end coordinates
paired_trips2
```

```python
paired_trips4_clean[paired_trips4_clean['start_time'] > paired_trips4_clean['end_time'] ]
```

```python
# Create downtime records: between end_time[i] and start_time[i+1]
downtime_records = []
for i in range(len(paired_trips4_clean) - 1):
    idle_start = paired_trips4_clean.loc[i, 'end_time']
    idle_end = paired_trips4_clean.loc[i + 1, 'start_time']
    idle_duration = (idle_end - idle_start).total_seconds() / 3600  # hours
    if idle_duration > 0:
        downtime_records.append({
            'idle_start_time': idle_start,
            'idle_end_time': idle_end,
            'idle_duration_hr': idle_duration,
            'idle_location_x': paired_trips4_clean.loc[i, 'end_x'],
            'idle_location_y': paired_trips4_clean.loc[i, 'end_y']
        })

# Convert to DataFrame
downtime_df = pd.DataFrame(downtime_records)
downtime_df.head(5)
```

```python
len(downtime_df['idle_start_time'])
```

```python
downtime_df.sort_values('idle_duration_hr')
```

```python
# Change from UTC to MDT
# Convert from UTC to MDT (Mountain Daylight Time, UTC-6)
downtime_df['idle_start_time_mdt'] = downtime_df['idle_start_time'].dt.tz_convert('America/Edmonton')
downtime_df['idle_end_time_mdt'] = downtime_df['idle_end_time'].dt.tz_convert('America/Edmonton')
```

```python
downtime_df
```

```python

# Add hour of day
downtime_df['idle_start_hour_mdt'] = downtime_df['idle_start_time_mdt'].dt.hour + \
                                 downtime_df['idle_end_time_mdt'].dt.minute / 60
```
